```python
import pandas as pd
```

```python
def createdf(gamename):
    # Creating df for each concurrent streams, concurrent viewers, hours watched
    exec(f"df_{gamename}_stream = pd.read_csv('{gamename}_concurrent_streams.csv')")
    exec(f"df_{gamename}_views = pd.read_csv('{gamename}_concurrent_viewers.csv')")
    exec(f"df_{gamename}_hours = pd.read_csv('{gamename}_hours_watched.csv')")
    
    # Renaming the columns based on where the data is coming from
    exec(f"df_{gamename}_stream = df_{gamename}_stream.rename(columns = {{'Average': 'Average_Stream', 'Gain':'Gain_Stream', '% Gain': '% Gain_Stream', 'Peak':'Peak_Stream'}})")
    exec(f"df_{gamename}_views = df_{gamename}_views.rename(columns = {{'Average': 'Average_Views', 'Gain':'Gain_Views', '% Gain': '% Gain_Views', 'Peak':'Peak_Views'}})")
    exec(f"df_{gamename}_hours = df_{gamename}_hours.rename(columns = {{'Average': 'Average_Hours', 'Gain':'Gain_Hours', '% Gain': '% Gain_Hours', 'Peak':'Peak_Hours'}})")
    
    # Creating the complete dataframe for the particular game
    exec(f"df_{gamename}_complete = df_{gamename}_stream.merge(df_cod_views, on = 'Month')")
    exec(f"df_{gamename}_complete = df_{gamename}_complete.merge(df_cod_hours, on = 'Month')")
    
    return
         
```

```python
df_cod_stream = pd.read_csv('cod_concurrent_streams.csv')
df_cod_views = pd.read_csv('cod_concurrent_viewers.csv')
df_cod_hours = pd.read_csv('cod_hours_watched.csv')

df_cod_stream = df_cod_stream.rename(columns = {'Average': 'Average_Stream', 'Gain':'Gain_Stream', '% Gain': '% Gain_Stream', 'Peak':'Peak_Stream'})
df_cod_views = df_cod_views.rename(columns = {'Average': 'Average_Views', 'Gain':'Gain_Views', '% Gain': '% Gain_Views', 'Peak':'Peak_Views'})
df_cod_hours = df_cod_hours.rename(columns = {'Average': 'Average_Hours', 'Gain':'Gain_Hours', '% Gain': '% Gain_Hours', 'Peak':'Peak_Hours'})

df_cod_complete = df_cod_stream.merge(df_cod_views, on = 'Month')
df_cod_complete = df_cod_complete.merge(df_cod_hours, on = 'Month')

df_cod_complete
```
