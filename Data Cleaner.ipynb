```python
import pandas as pd
```

```python
df_lol = pd.read_csv('Data/combined_twitchtracker_lol_clean.csv')
```

```python
def cleandate(df):
    for index, row in df.iterrows():
        x = row['Month'].split('-')
        newmonth = f'{x[1]}-{x[0]}'
        df.at[index,'Month'] = newmonth
        
    return df
```

```python
df_lol_clean = cleandate(df_lol)
df_lol_clean
```

```python
df_fortnite = pd.read_csv('fortnite_concurrent_viewers.csv')
df_fortnite_clean = cleandate(df_fortnite)
```

```python
df_fortnite_clean.to_csv('fortnite_concurrent_viewers.csv', index = False)
```
